```python
def get_rosalind_data(filepath, filename, readlines=False):    
    
    if readlines:
        with open(filepath+filename, "r") as f:
            data = f.readlines()
            data = [line.strip("\n") for line in data]
    else: 
        with open(filepath+filename, "r") as f:
            data = f.read().strip("\n")

    return data

filepath="C:\\Users\\adeoy\\Downloads\\"
filename="rosalind_revp (1).txt"

polypeptide = "".join([item for item in get_rosalind_data(filepath, filename, readlines=True)[1:]])
```

```python
palin_length = [i for i in range(4, 13)]

for length in palin_length:
    for i in range(len(polypeptide)-length+1):
        reverse_comp = "".join(["T" if base=="A" else "C" if base=="G" else "G" if base=="C"
                            else "A" if base=="T" else "" for base in polypeptide[i:i+length]])[::-1]
        if reverse_comp == polypeptide[i:i+length]:
            print(i+1, length)
    
```
